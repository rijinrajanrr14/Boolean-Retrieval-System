```python
#doc_path=input("Give the path of document:")
def Document(*doc_path):
    doc_list=[]
    for p in doc_path:
        df=''
        df=open(p,"r")
        i=df.read()
        doc_list.append(i)
    return doc_list
```

```python
#Removing some undesired characters from our output doc
def cleaner(a=[]):
    k='''\xa0'''
    c=""
    for i in a:
        if i not in k:
            c=c+i
    return c
```

```python
#Removing punctuations
def punctuation_removal(str):
    no_punct=""
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for char in str:
        if char not in punctuations:
            no_punct=no_punct+char
            
    return no_punct
    
```

```python
#Tokenizing the string 
def tokenizer(str):
    words=[]
    words.append(str.split())
    return words
```

```python
import nltk
def stopwords(tokens):
    from nltk.corpus import stopwords
    nltk.download('stopwords')
    stop_words = stopwords.words('english')
    txt=[]
    output=[]
    for li in tokens:
        for i in range(0,len(li)):
            word=li[i]
            word=word.lower()
            if word not in stop_words:
                output.append(word)
            
    return output
```
