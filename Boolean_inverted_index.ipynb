```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
def docs(*doc_path):
    doc_lst=[]
    for p in doc_path:
        data= open(p)
        y=data.read()
        
        punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        for ele in y:
            if ele in punc:
                y = y.replace(ele, "")
        
        tokens=[t for t in y.split()]
        
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        sr= stopwords.words("english")
        clean_tokens=[]
        for token in tokens:
            
            wd=token.lower()
            if wd not in sr:
                clean_tokens.append(wd)
        
        doc_lst.append(clean_tokens)
    return doc_lst
```
